
## 简介

为了提供如何应用[数据模型](/model/)的示例，本节是一个社区贡献的烹饪指南，包含了将不同成分组合在一起以产生期望结果的食谱。每个贡献组织都有自己的部分，他们可以在其中提供尽可能多的不同食谱，只要有时间即可。食谱在提交时会经过适当性和正确性检查。
