
## 示例

本节包含使用 Linked Art 数据模型的示例。这些示例展示了如何为不同类型的艺术品和文化遗产项目建模。

## 示例类型

* **艺术品示例**：展示如何描述绘画、雕塑、照片等
* **人物示例**：展示如何描述艺术家、收藏家、策展人等
* **机构示例**：展示如何描述博物馆、画廊、学术机构等
* **事件示例**：展示如何描述展览、拍卖、创作过程等
* **地点示例**：展示如何描述博物馆、工作室、展览场所等

每个示例都包含完整的 JSON-LD 表示，以及关于建模选择的解释。
