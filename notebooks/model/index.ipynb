## 简介

Linked Art 数据模型是一个应用纲要，可用于描述文化遗产资源，重点关注艺术品和面向博物馆的活动。它定义了通用模式和术语，以确保生成的数据能够被轻松[使用](/loud/)，并且基于真实世界的数据和用例。

1.0 版本的模型开发已经完成，文档正在整理以备不久后的正式发布。所有决策都已做出并记录，剩下的工作就是检查一致性和准确性。然而，这并不意味着工作停止——社区将继续随时间改进模型，并已确定需要在未来版本中讨论的问题。

## 模型开发

### 模型背景

遵循社区现有的标准和最佳实践，我们的起点包含以下背景标准：

* [CIDOC-CRM](https://www.cidoc-crm.org/) 作为核心概念模型，提供基于事件的范式。
    * 我们使用 [CIDOC-CRM 7.1.3](https://www.cidoc-crm.org/Version/version-7.1.3) 的简化[子集](profile/)以确保一致性和可理解性。
* [Getty 词表](http://vocab.getty.edu/) 作为领域特定术语的核心身份来源
    * 术语分为三个列表：[必需的](vocab/required/)、[推荐的](vocab/recommended/)和[可选的](vocab/optional/)，基于如果做出不同选择系统是否仍能互操作。
* 纲要的具体表达和数据访问在 [API](/api/) 文档中全面涵盖。
    * 核心序列化格式是 [JSON-LD 1.1](https://w3.org/TR/json-ld11)


### 范围和流程

#### 开发流程

艺术领域关联开放数据的理想目标模型应具有以下属性：

* 尽可能多地捕获我们了解到的关于资源的信息
* 可以通过易于实现的[服务](/api/)有效使用
* 提供与其他相关数据集的互操作性
* 解决实际挑战，这些挑战被记录为用例

成功的模型开发遵循以下原则：

* 迭代式（我们不会一次就做对）
* 响应式（我们将根据反馈和关注点修改模型）
* 负责任地（我们将谨慎考虑变更和功能，权衡复杂性和价值）
* 协作式（我们将经常并尽早与社区、项目和个人接触）

有关如何参与 Linked Art 的更多信息，请参阅网站的[社区](/community/)部分。

#### 范围限制

* 互操作性
    * 工作的目标是系统之间的互操作性，而不是提供全面的数据模型来描述单个机构可能知道的所有信息。这意味着数据的管理和生产超出了范围，作为系统特定的关注点。同样，这些系统内使用的确切模型也不是我们的关注点。否则会违反技术独立性原则。开放访问不是必需的，但意图在系统边界之外发布是必需的。
* 完整的书目和档案描述
    * 在关联数据中彻底描述书目和档案资源是图书馆和档案领域正在进行的几个讨论的主题。Linked Art 通过使用我们的共享模式定义基本描述为这些领域提供入口，但不会试图在这些领域做到完整。
* 数据来源
    * 记录数据本身被创建、修改和管理的个别事件超出了本工作的范围。全球关联开放数据社区有各种方法来解决这个问题，具有不同程度的复杂性和准确性。考虑到艺术领域工作的相对不成熟，我们认为在整个数据集以下的任何级别添加此内容都是不必要的负担。
* 不确定性和信念的量化
    * 同样，对所表达数据的确定性程度或信念对研究人员来说很有价值且令人感兴趣，但需要显著更复杂的环境。这也将被证明是不可持续的负担，即使提供了也不切实际。
