
## 概述

任何艺术史信息系统的核心都是艺术对象本身。在人类社会中具有某些价值的物理对象的类是 `HumanMadeObject`（人造物）。即使对象是一个贝壳（因此是自然的）且没有任何重大修改，这个类仍然被使用，因为修改是指社会赋予其价值。这也使得模型易于使用，而不必担心那些对区分具有重大意义的改变程度。
