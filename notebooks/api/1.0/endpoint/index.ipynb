

## 概述

Linked Art API 由不同的端点组成，每个端点都有定义的返回数据格式结构。这些（主要）与模型的核心类对齐，并根据[API 设计原则](../principles/)构建。

## 端点

* [抽象作品](abstract_work/) - 抽象作品，如展览的想法
* [概念](concept/) - 类型、材料、语言等，作为完整记录而不是外部引用
* [数字对象](digital_object/) - 图像、服务和其他数字对象
* [事件](event/) - 相关但不是其他实体一部分的事件和其他非特定活动
* [群体](group/) - 群体和组织
* [个人](person/) - 个人
* [物理对象](physical_object/) - 物理事物，包括艺术品、建筑物或其他建筑、书籍、对象部分等
* [地点](place/) - 地理地点
* [流传历史活动](provenance_activity/) - 物理事物历史期间发生的各种事件
* [集合](set/) - 集合，包括收藏集和用于展览的对象集合
* [文本作品](textual_work/) - 值得作为独立实体描述的文本，如书籍或期刊文章承载的内容
* [视觉作品](visual_work/) - 值得作为独立实体描述的图像内容，如绘画或素描显示的图像
