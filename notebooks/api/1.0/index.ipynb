## 概述

模型或应用纲要定义了 Linked Art 使用的概念模型、本体和词汇表术语的方面。它不是相对于任何特定的技术交互方法定义的，用于检索、更新、收割、搜索或浏览，允许许多不同的可能实现。

网站的本节记录了基于 Web 的 API 或应用程序编程接口，应用程序可以通过它与来自符合发布者的数据交互。此 API 在设计时考虑了几个设计原则，以确保对软件开发者尽可能可用。文档分为几个部分，将 HTTP 上的协议交互与关于返回 JSON 结构和不同 JSON 端点之间链接的格式决策分开。端点按模型的核心实体分割，共享结构如名称和标识符被单独描述。


## Linked Art API

### 核心实体描述 API

核心 Linked Art API 是提供描述遵循模型的实体的记录，根据 Linked Art 上下文文档序列化为 JSON-LD。响应还有一个非语义块，遵循超文本应用语言（或 HAL）规范，以提供版本、搜索和其他相关信息的链接。

* [实体描述端点](endpoint/) - 将对象、作品、个人、地点和其他实体检索为 JSON-LD
    * [共享数据结构](shared/) - 多个描述使用的共享模式
    * [JSON 模式](schema_docs/) - API 的 JSON 模式定义，附带文档
* [超文本应用语言](hal/) - 版本控制、跨格式链接和其他 HAL 链接


### 搜索 API

Linked Art 定义了搜索请求响应的格式，但目前没有定义用于定义搜索的查询语言。标准化响应的预期用途是查询可以通过记录内的 HAL 链接引用，消费应用程序在跟随链接时将收到一致的格式。单个实现可能使用 [SPARQL](https://www.w3.org/TR/sparql11-query/)、[GraphQL](https://graphql.org/)、Cypher、Elasticsearch DSL、Solr 查询语言或其他可以嵌入 URL 的语法，而应用程序无需知道如何构造它。

* [搜索 API](search/) - Linked Art 搜索的响应格式，以及如何使用 HAL 链接到它们
* [搜索链接](../rels/1/) - HAL 链接的命名搜索列表


### 数据发现 API

在数据意图是开放可用以供使用和重用的环境中，能够将数据收割到本地系统进行处理和索引是很重要的。随后，能够与数据保持同步，知道记录何时被创建、更改或删除是很重要的。对于应用程序来说，知道用于生成 Web 界面的数据在哪里也很有用，特别是对于浏览器内应用程序。

* [发现 API](discovery/) - 使记录可用于轻松收割和从 HTML 发现

### API 版本控制

我们遵循[语义版本控制](https://semver.org/spec/v2.0.0.html)的概念，版本号结构为 _Major.Minor.Patch_。所有次要版本与其主版本向后兼容，但可能添加新功能或澄清未明确定义的技术方面。补丁版本仅用于不影响功能的文档更改，将在原地进行而不更改文档的 URI。主要版本保留用于向后不兼容的更改，并将谨慎进行。
