这些数据结构在许多端点之间共享，在主响应内。

## 数据结构

* [活动](activity/) - 嵌入在主实体记录中的创作、出版、出生/死亡和其他活动
* [数字链接](digital/) - 到其他数字内容的链接，如网页和图像
* [尺寸](dimension/) - 资源的尺寸，具有给定单位和类型（例如，20英寸宽）
* [标识符](identifier/) - 资源的标识符（例如，1997-1234的入藏号）
* [货币金额](money/) - 给定货币的货币金额（例如，10000美元）
* [名称](name/) - 已给予资源的人类语言名称（例如，英文的《蒙娜丽莎》）
* [权利](right/) - 许可证或财产权的结构化权利断言
* [引用](reference/) - 对系统中另一个实体的引用（例如，从绘画到艺术家信息的引用）
* [陈述](statement/) - 描述资源的人类可读文本（例如，材料陈述）
* [时间范围](timespan/) - 时间跨度，两端都有模糊性（例如，1950年到1955年）
* [类型 / 概念](type/) - 进一步澄清实体的类型引用（例如，绘画而不是对象）
* [关系](assignment/) - 实体之间的关系分配（例如，相关作品、家庭关系）


## 上下文和URI

请注意，嵌入在端点可用记录中的数据结构没有自己的`@context`条目，因为它们使用主记录的上下文文档。如果数据结构使用linked art格式单独共享，那么它必须添加`@context`引用。

如果有一个数据结构通常完全嵌入在另一个记录中，如尺寸或名称（不是对另一个记录的引用），在`id`中给出了自己的URI，可以从中检索数据，并且那里有关于实体的更多信息，那么linked art记录中的结构必须有一个名为`_complete`的属性，值为`false`。这将使数据消费者知道从`id`字段中给出的URI检索数据可能是有价值的。否则，`id`中的URI仅是信息性的。




```
# 导入 cromulent 库
from cromulent import model, vocab

top = model.HumanMadeObject(ident="auto int-per-segment")
n = vocab.PrimaryName(ident="auto int-per-segment", content="Hacha")
n.language = vocab.instances['spanish']
n._complete = False
top.identified_by = n


# 展示生成的 JSON-LD
print(model.factory.toString(top, compact=False))
```
