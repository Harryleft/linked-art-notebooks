
## 参考文献

这个精选收藏包含讨论或实施 Linked Art 模型、API 规范和社区倡议的学术出版物、书籍章节、技术报告和研究输出。

条目按时间顺序排列（从最新到最旧），然后在每年内按作者姓氏的字母顺序排列。请注意，Linked Art 模型随着时间的推移显著演变，1.0 版本于 2025 年 2 月正式发布。虽然模型在 2024 年全年相对稳定，变化较小，但 2024 年之前的出版物可能引用模型在形成阶段的早期开发迭代，在标准化完成之前。

要建议向此参考文献添加内容，请在 GitHub 上提交拉取请求或打开问题，在 Linked Art Slack 工作区发送消息，或发布到 [Linked Art Google Group](https://groups.google.com/g/linked-art)。
