
## 软件

这是一个页面，包含让您实现所有这些内容的软件引用。

## 库

* [Crom](http://github.com/linked-art/crom)
> 一个 Python 库，试图使将资源视为具有属性的类实例而不是图更容易。
* [LinkedArt.js](https://linkedartjs.org/)
> 处理 linked art JSON 的 Javascript 库
* [Linked.Art.Net](https://github.com/tomcrane/linked-art-net)
> 用于 Linked Art 的 C# / DotNet 库


## 平台

* [LUX](https://github.com/project-lux/)
> [LUX](https://lux.collections.yale.edu/) 的开源实现，耶鲁基于 Linked Art 的跨收藏发现平台
* [Arches](https://www.archesproject.org/)
> 一个文化遗产关联数据管理平台，适合实施 Linked Art
* [Ogee](https://github.com/thegetty/Ogee)
> Ogee Arches 是为 Arches 平台设计的包，实现了 Linked Art 数据模型（截至2023年12月），提供完整的词汇表来支持该数据模型，并具有可重用文档的完整文档。


## 文档管理

* [Zellij](https://zellij.pythonanywhere.com/docs/list/apppWYuo1z7E2J7E8?flag=0)
> Zellij 是一个语义数据模式库，提供可重用的语义数据模式库，可用作语义数据建模项目的范式。Zellij 上的 Linked Art 语义数据模型库以组合和用户友好的格式表示 Linked Art 模式。这种表示的目的是帮助标准的潜在用户学习和应用它，包括为语义数据管理平台生成派生数据管理产品，如查询、映射模板和显示模板。
