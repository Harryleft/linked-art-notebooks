
## 概述

[关联开放数据](http://5stardata.info/en/)以其五星卓越标准改变了我们在网络上发布数据的方式。它给了我们一个简短且非常实用的检查清单，我们可以用它来从考虑语义转向发布实际数据。它促进开放性、标准和系统之间的链接作为重用的必要条件。

但它完全来自发布视角，没有就如何以潜在消费者可用的方式发布数据做出任何建议。在此期间，Web 社区认识到我们需要另外5颗星或设计原则来促进数据消费。如果我们的数据不被使用，那么投资于其创建、发布、维护和改进的资源就没有获得任何价值。如果我们希望我们的数据被使用，那么它需要是**可用的**：关联开放可用数据。

## 可用性

根据维基百科，

> ... *可用性* 是指定消费者可以在量化的使用情境中有效地、高效地、满意地使用[某物]来实现[他们]量化目标的程度。

这意味着可用性依赖于受众。数据的受众是开发者，而开发者的用户界面是他们访问数据的 API。LOD 的 API 通过 HTTP，这是很好理解的，但也依赖于本体。本体主要是为了语义或理论正确性而设计的，这是最不实际的关注点。我们需要平衡表达完整性和精确性与结果数据构造的可用性之间的权衡。

JSON-LD 允许将本体构造映射到 JSON，现代开发者的通用语言，并且是 LOUD 的基石技术。五个设计原则是可用性的基线。


## LOUD 的设计原则

### A. 适合受众的正确**抽**象

开发者不需要与本体学家相同的数据访问级别，就像司机不需要与技师相同级别的汽车内部工作访问一样。用例和需求应该驱动系统之间的互操作性层，而不是本体纯度。

### B. **低**入门门槛

开始使用数据并构建东西应该很容易。如果需要很长时间来理解模型、本体、sparql 查询语法等等，那么开发者会寻找更容易的目标。相反，如果容易开始并逐步改进，那么更多人会使用数据。

### C. 通过内省**可**理解

数据应该在很大程度上仅仅通过查看就能理解，而不是要求开发者阅读本体和词汇表。使用 JSON-LD 让我们用开发者已经理解的语言与他们交谈。

### D. 带有工作示例的**文**档

你永远无法凭直觉了解数据的所有规则。文档阐明了开发者可以期望遇到的模式，使他们能够稳健地实现。示例用例允许在何时遇到模式进行上下文化，工作示例让您可以将数据放入系统以查看它是否正确实现了该模式。

### E. **少**例外，而是许多一致的模式

你在 API（因此也是本体）中拥有的每个例外都是开发者需要学习的另一个规则。每个例外都是突兀的，需要额外的代码来管理。虽然不是一切都是同质的，但一组能很好地管理例外的模式比许多自定义字段更好。
