
## 什么是 Linked Art？

__Linked Art 是一种基于标准的可用方法，用于描述文化遗产。__

或者，更详细地说，Linked Art 是一个由博物馆和文化遗产专业人士组成的社区，协作定义用于描述文化遗产的元数据应用纲要（[模型](/model/)），以及与之便捷交互的技术手段（[API](/api/)）。我们举办线下[活动](/community/events/)以协助学习、使用和改进 Linked Art。

模型和 API 基于适当的国际标准，以确保它们具有互操作性且易于使用。我们重点关注数据的可用性，使其易于实现和维护。我们开发服务以协助实施，但不维护商业产品。
