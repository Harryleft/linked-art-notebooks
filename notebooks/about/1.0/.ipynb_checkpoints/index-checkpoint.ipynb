

**2025年2月19日**

Linked Art 社区激动地宣布 Linked Art 1.0 规范的正式发布，这是一个重要的里程碑，为文化遗产部门提供了共享和连接收藏品及其物品历史背景信息的标准方法。Linked Art 已被博物馆部门的主要组织采用，提供了一个基于现有国际标准构建的稳健且经过验证的数据框架，用于表示艺术品、档案、参考资料、艺术家、地点、组织、概念和事件（如展览或来源信息）之间的丰富关系。

通过采用 Linked Art，文化机构将增强其收藏品的发现性和可访问性，实现更有吸引力的界面，并通过遵循记录中表达的关系使搜索引擎更容易发现和抓取收藏品。这些关系能够表达更精确的描述，使知识系统能够理解和处理，实现下一代语义搜索，并促进使用收藏品进行深度研究。使用 Linked Art，耶鲁大学首次能够将其博物馆、图书馆和档案馆的收藏连接在单一发现环境中，通过使主要搜索引擎能够抓取知识，每月页面浏览量从数千增加到数十万。

耶鲁大学数字文化遗产高级总监、Linked Art 编辑委员会联合主席 Robert Sanderson 博士将 1.0 的发布描述为"文化遗产标准世界的关键时刻。国际图像互操作性框架（<a href="https://iiif.io/">IIIF</a>）彻底改变了我们参与基于图像的数字内容的方式，Linked Art 准备从根本上改变我们处理关于这些图像中描绘对象的知识的方式。我们在耶鲁成功构建 LUX（我们的图书馆、档案馆和博物馆的跨收藏发现环境）如果没有 Linked Art 社区过去七年的努力工作是不可能实现的。"

<image src="https://images.collections.yale.edu/iiif/2/ycba:4f227f08-7842-46cc-b05a-e3c6a4614cc1/full/800,/0/default.jpg"><br/>
<i><a href="https://lux.collections.yale.edu/view/object/047181f4-2fdf-43fd-abbe-f16d7f9edc10">多特或多德雷赫特：从鹿特丹来的多特邮船被围困</a>，作者 J. M. W. 特纳，藏于耶鲁英国艺术中心</i>

Linked Art 1.0 反映了数十个组织尝试描述其收藏品，然后构建和部署系统的经验，如耶鲁的 LUX 和盖蒂的研究收藏查看器，这些系统提供了对开发者友好、可互操作和跨领域的解决方案。"我们的软件工程师能够快速理解要求，因为标准一致且描述清晰，即使他们没有文化遗产背景。"桑德森在谈到开发经验时说。

"盖蒂已采用 Linked Art 作为其档案和博物馆收藏的标准数据模型和 API，通过提供一致的方式来理解文化物品如何在时空中移动，赋能公众和我们的技术收藏工作人员。我们也很高兴 Linked Art 使用盖蒂的艺术与建筑词表（AAT）词汇表作为核心组件来实现跨组织的互操作性，有助于实现我们的使命，并展示我们二十多年来投资标准的价值。"J·保罗·盖蒂信托公共技术高级总监 David Newbury 说。

根据 Newbury 的说法，Linked Art 既提供了描述文化遗产知识的标准模型，确保意义的一致性，也提供了标准的 Web 应用程序编程接口（API），允许数字系统以直观的方式与该知识交互。这个第一个稳定版本标志着遗产部门在促进跨机构和世界各地的互操作性和知识共享方面迈出了重要一步。Linked Art 已准备好被广泛采用，使用现有的免费开源软件或在准备好帮助的供应商协助下。盖蒂并不是唯一实施它的博物馆，华盛顿特区的国家艺术馆和阿姆斯特丹的国家博物馆都深度参与其中。

<image src="https://media.getty.edu/iiif/image/8c255d80-7382-46db-9fa8-892c0d37247e/full/800,/0/default.jpg"><br/>
<i><a href="https://www.getty.edu/art/collection/object/103JNH">鸢尾花</a>，作者文森特·梵高，藏于 J·保罗·盖蒂博物馆</i>

"Linked Art 代表了国家画廊数字战略的变革性一步，使我们能够以以前难以想象的方式将我们的世界级收藏带给观众。通过利用 Linked Art 开放标准，我们不仅增强了我们收藏品的可发现性——我们从根本上重新思考了我们如何在时空中连接艺术品、艺术家和展览。这种合作直接支持国家画廊通过促进对艺术及其历史的更深层理解来为国家服务的使命，使我们共享的文化遗产对所有人更加可及。这是一个激动人心的时刻，我们很自豪站在这一创新的前沿。" – Nick Sharp，国家画廊首席数字官

"Linked Art 是支持国家博物馆新 Collectie Online 平台的核心技术之一，这是一个探索荷兰艺术和历史的创新界面。Linked Art 使我们能够分享关于一百万博物馆对象的知识作为一个丰富互联的语义数据网络，为我们的信息和数据的新互动提供动力。通过围绕社区需求构建的数据模型，以及为 Web 开发者优化的 API，Linked Art 为文化遗产数据发布带来了稳定性和可持续性。" – Saskia Scheltjens，国家博物馆研究服务主管和首席图书馆员

这项工作也得到了行业的支持，有几家公司提供实施规范和支持工作的服务和产品。"根据我们的经验，决定使用和创建语义数据的机构和研究人员希望建立在既定实践的坚实基础上。使用 Linked Art 作为标准，我们能够支持我们的客户实施社区创建和维护的语义数据模式来处理文化遗产中许多最常见的数据建模情况。Linked Art 为我们和我们的客户节省了数百小时不必要的基础工作，这些工作已经完成，所以我们可以推动语义数据表达的边界前进，并将创建关联知识网络的目标向前推进。"Takin.Solutions 首席执行官 George Bruseker 写道。

Bruseker 应该知道，因为他也是国际博物馆理事会（ICOM）概念参考模型组的联合主席，Linked Art 就是从中衍生出来的。

对于大学和博物馆来说，Linked Art 也激活了收藏用于研究，推进了最近"收藏作为数据"的趋势。通过指定标准的描述模式、标准的交互方式，然后允许包含额外的领域或收藏特定知识，Linked Art 消除了当前跨收藏和跨机构研究的障碍。这使得汇集来自世界各地的相关信息成为简单的第一步，而不是为每个新研究项目繁琐地重新实施复杂的解决方案。

"Linked Art 是扩展人文学科数字学术研究的重要基础，开辟了使用我们期待在未来研究中扩展的前沿数字方法进行跨收藏研究的潜力。1.0 版本代表了一个重要的里程碑，收藏数据现在可从世界上越来越多的艺术机构获得。我们很高兴牛津大学电子研究中心能够在过去六年中为这一社区努力做出贡献，这要感谢英国艺术与人文研究理事会的关键支持。"牛津大学电子研究中心高级研究员和兼职教员 Kevin Page 博士说。


### 关于 LINKED ART：

Linked Art 是一个持续的合作努力，致力于开发、推广和支持基于关联开放可用数据的原则和技术构建的标准数据模型和 API，用于描述文化和自然历史及其历史和机构背景。充满活力的社区由文化遗产专业人员、技术专家和研究人员组成。
