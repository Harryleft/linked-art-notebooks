## 参与方式

有很多参与方式：

* 参加 Linked Art [活动](/community/events/)
* 加入讨论组：<https://groups.google.com/forum/#!forum/linked-art>
* 加入 Slack 工作区（发送邮件至 [robert.sanderson@yale.edu](mailto:robert.sanderson@yale.edu) 获取邀请）
* 添加问题/疑问：<https://github.com/linked-art/linked.art/issues>
* 在 Github 中编辑网站源代码：<https://github.com/linked-art/linked.art/>

## 资助项目和支持者

Linked Art 倡议直接和间接地得到多个项目及其资助者的支持：

* [Linked Art：网络化数字收藏和学术研究](/community/projects/researchnetwork/)（项目参考 AH/S012419/1）一个由英国[艺术与人文研究理事会](https://ahrc.ukri.org/)（AHRC）资助的、基于牛津大学的研究网络
* [Linked Art II：发展社区、实践和学术研究](/community/projects/linkedartii/)（项目参考 AH/T013117/1），同样由英国[AHRC](https://ahrc.ukri.org/)在牛津大学资助
* [塞缪尔·H·克雷斯基金会](http://www.kressfoundation.org)的拨款，支持编辑委员会的工作
* [丰富展览故事：为 Quire 添加声音](/community/projects/ees2)（项目参考 AH/Y006011/1）由英国[艺术与人文研究理事会](https://ahrc.ukri.org/)（AHRC）资助，基于牛津大学
