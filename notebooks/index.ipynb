
Linked Art 是一个[社区](/community/)，致力于共同创建基于[关联开放可用数据](/loud/)的共享[模型](/model/)，用于描述文化遗产，重点关注艺术品，同时也包括档案和书目材料。我们制定了能够访问模型中数据的 [API](/api/)，并在[软件](/software/)中实现它们，为世界提供有价值的内容。
